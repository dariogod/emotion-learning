```python
import json

with open("outputs/bert-without-emotion_2/checkpoint-1000/trainer_state.json", "r") as f:
    trainer_state_without_emotion = json.load(f)

with open("outputs/bert-with-emotion_2/checkpoint-1000/trainer_state.json", "r") as f:
    trainer_state_with_emotion = json.load(f)
```

```python
import matplotlib.pyplot as plt

# Extract evaluation accuracies and steps
eval_steps_without = []
eval_acc_without = []
for log in trainer_state_without_emotion['log_history']:
    if 'eval_accuracy' in log:
        eval_steps_without.append(log['step'])
        eval_acc_without.append(log['eval_accuracy'])

eval_steps_with = []
eval_acc_with = []
for log in trainer_state_with_emotion['log_history']:
    if 'eval_accuracy' in log:
        eval_steps_with.append(log['step'])
        eval_acc_with.append(log['eval_accuracy'])

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(eval_steps_without, eval_acc_without, label='Without Emotion', marker='o')
plt.plot(eval_steps_with, eval_acc_with, label='With Emotion', marker='o')

plt.xlabel('Training Steps')
plt.ylabel('Evaluation Accuracy')
plt.title('BERT Model Evaluation Accuracy During Training')
plt.legend()
plt.grid(True)
plt.show()
```
